```python
import os
from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
```

```python
import osmnx

graph = osmnx.graph_from_address(
    address="Argyle Street",
    network_type="drive_service",
    simplify=False,
)
osmnx.plot_graph(G=graph)
```

```python
gdf_nodes, gdf_edges = osmnx.graph_to_gdfs(G=graph)
```

```python
gdf_nodes
```

```python
gdf_edges
```

```python
import pandas as pd

nodes_df = pd.DataFrame(data=gdf_nodes)
nodes_df
```

```python
nodes_df.to_csv(path_or_buf=f"{ROOT_DIR}/data/raw/osm_nodes_info.csv")
```

```python
import pandas as pd

edges_df = pd.DataFrame(data=gdf_edges)
edges_df
```

```python
edges_df.to_csv(path_or_buf=f"{ROOT_DIR}/data/raw/osm_edges_info.csv")
```
