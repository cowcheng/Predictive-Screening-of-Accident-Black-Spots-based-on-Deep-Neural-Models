```python
import os
from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
```

```python
from datetime import datetime, timedelta

import requests
from tqdm import tqdm
from bs4 import BeautifulSoup

url_template = "https://programme.rthk.hk/channel/radio/trafficnews/index.php?d={date}"
start_date = datetime(
    year=2023,
    month=1,
    day=1,
)
news_info_list = []
for i in tqdm(iterable=range(0, 10)):
    news_date = (start_date + timedelta(days=i)).strftime(format="%Y%m%d")
    response = requests.get(url=url_template.format(date=news_date))
    html = BeautifulSoup(
        markup=response.content,
        features="html.parser",
    )
    articles = html.find(
        name="div",
        attrs={"class": "articles"},
    )
    inner = articles.find_all(
        name="li",
        attrs={"class": "inner"},
    )
    if inner:
        for info in inner:
            data = info.get_text().split("\t")
            date = data[-1].strip().split(" HKT ")[0]
            time = data[-1].strip().split(" HKT ")[1]
            detail = data[0]
            news_info = [date, time, detail]
            if news_info not in news_info_list:
                news_info_list.append(news_info)
news_info_list
```

```python
import pandas as pd

news_df = pd.DataFrame(
    data=news_info_list,
    columns=["date", "time", "detail"],
)
news_df
```

```python
news_df.to_csv(path_or_buf=f"{ROOT_DIR}/data/raw/traffic_news_info_sample.csv")
```
