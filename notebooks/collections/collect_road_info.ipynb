# Collect road information

```python
import json

import pandas
import requests
```

```python
url = 'https://www.overview.hk/street/ssp.php'

resp = requests.get(url=url)
resp_dat = json.loads(s=resp.text)

resp_dat
```

```python
road_prop = {
    'Name(Chinese)': [],
    'Name(English)': [],
    'Location(Chinese)': [],
    'Location(English)': [],
    'AdditionalInfo': [],
}

for data in resp_dat['data']:
    road_prop['Name(Chinese)'].append(str(data[0]))
    road_prop['Name(English)'].append(str(data[1]))
    road_prop['Location(Chinese)'].append(str(data[2]))
    road_prop['Location(English)'].append(str(data[3]))
    road_prop['AdditionalInfo'].append(str(data[4]))

road_prop
```

```python
road_df = pandas.DataFrame(data=road_prop)

road_df
```

```python
road_df.to_csv(path_or_buf='../../datasets/raw/road_info.csv', index=False)
```
