# Collect traffic news information

```python
import datetime

import pandas
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://programme.rthk.hk/channel/radio/trafficnews/index.php?d={date}'
news_prop = {
    'Message': [],
    'Datetime': [],
}

start_date = datetime.datetime(year=2015, month=1, day=1)

for i in range(0, 365):
    news_date = (start_date + datetime.timedelta(days=i)).strftime(format='%Y%m%d')
    resp = requests.get(url=url.format(date=news_date))
    soup = BeautifulSoup(markup=resp.content, features='html.parser')
    articles = soup.find(name='div', attrs={'class': 'articles'})
    ul = articles.find_all(name='li', attrs={'class': 'inner'})
    if i % 10 == 0:
        print(news_date)
    if ul:
        for li in ul:
            news = li.get_text().split(sep='\t')
            news_msg = news[0]
            news_datetime = news[-1]
            news_prop['Message'].append(str(news_msg))
            news_prop['Datetime'].append(str(news_datetime).strip())

news_prop
```

```python
news_df = pandas.DataFrame(data=news_prop)

news_df
```

```python
news_df.to_csv(path_or_buf='../../datasets/raw/traffic_news_info.csv', index=False)
```
