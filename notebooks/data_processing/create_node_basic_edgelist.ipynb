```python
import os
from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
```

```python
import pandas as pd

edges_df = pd.read_csv(filepath_or_buffer=f"{ROOT_DIR}/data/raw/osm_edges_info.csv")
edges_df
```

```python
edges_df = edges_df.loc[:, ["u", "v"]]
edges_df
```

```python
edges_list = edges_df.values.tolist()
edges_list
```

```python
with open(f"{ROOT_DIR}/data/training/node2vec/basic.el", "w") as f:
    for edge in edges_list:
        f.write(f"{edge[0]} {edge[1]}\n")
```
