```python
import os
from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
```

```python
import pandas as pd

edges_df = pd.read_csv(filepath_or_buffer=f"{ROOT_DIR}/data/raw/osm_edges_info.csv")
edges_df
```

```python
edges_df = edges_df.loc[:, ["u", "v", "ref"]]
edges_df
```

```python
edges_list = edges_df.values.tolist()
for edge in edges_list:
    edge[0] = int(edge[0])
    edge[1] = int(edge[1])
    if str(edge[2]) == "nan":
        edge[2] = -1.0
edges_list
```

```python
with open(f"{ROOT_DIR}/data/training/node2vec/ref.el", "w") as f:
    for edge in edges_list:
        f.write(f"{edge[0]} {edge[1]} {edge[2]}\n")
```
