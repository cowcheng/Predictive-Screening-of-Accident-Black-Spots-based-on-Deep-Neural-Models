```python
import os
from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
```

```python
import pandas as pd

edges_df = pd.read_csv(filepath_or_buffer=f"{ROOT_DIR}/data/raw/osm_edges_info.csv")
edges_df
```

```python
edges_df = edges_df.loc[:, ["u", "v", "oneway"]]
edges_df
```

```python
oneway_dict = edges_df["oneway"].value_counts().to_dict()
oneway_dict
```

```python
oneway_dict_keys_list = list(oneway_dict.keys())
oneway_dict_keys_list
```

```python
oneway_value_dict = {
    category: index for index, category in enumerate(oneway_dict_keys_list)
}
oneway_value_dict
```

```python
edges_list = edges_df.values.tolist()
for edge in edges_list:
    edge[2] = float(oneway_value_dict[edge[2]])
edges_list
```

```python
with open(f"{ROOT_DIR}/data/training/node2vec/oneway.el", "w") as f:
    for edge in edges_list:
        f.write(f"{edge[0]} {edge[1]} {edge[2]}\n")
```
