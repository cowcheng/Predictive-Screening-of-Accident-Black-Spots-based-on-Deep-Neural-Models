```python
import os
from pathlib import Path

ROOT_DIR = Path(os.getcwd()).parent.parent.absolute().__str__()
```

```python
edgelist_file_path = f"{ROOT_DIR}/data/training/node2vec/tunnel.el"

dimensions = 128
walk_length = 80
num_walks = 10
```

```python
import multiprocessing

import networkx

workers = multiprocessing.cpu_count()
weight_key = Path(edgelist_file_path).name.split(".")[0]
graph = networkx.read_edgelist(
    path=edgelist_file_path,
    data=[(weight_key, float)],
)
graph
```

```python
from node2vec import Node2Vec

n2v_model = Node2Vec(
    graph=graph,
    dimensions=dimensions,
    walk_length=walk_length,
    num_walks=num_walks,
    weight_key=weight_key,
    workers=workers,
)
n2v_model
```

```python
word2vec_model = n2v_model.fit()
word2vec_model
```

```python
word2vec_model.wv.save_word2vec_format(
    fname=f"{ROOT_DIR}/models/node2vec/{weight_key}_{dimensions}-{walk_length}-{num_walks}.bin",
)
```
